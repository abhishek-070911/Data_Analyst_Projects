# 2. Data Cleaning

```python
# Checking for missing values

ocd.isnull().sum()
```

```python
# Filling NaN values with diagnosis
ocd['Previous Diagnoses'] = ocd['Previous Diagnoses'].fillna('Unknown')
```

```python
# Filling NaN values with medications
ocd['Medications'] = ocd['Medications'].fillna('Unknown')
```

```python
# Rechecking for missing values

ocd.isnull().sum()
```

* There are no missing values present.

```python
# Converting Diagnosis date from object to datetime

ocd['OCD Diagnosis Date']= pd.to_datetime(ocd['OCD Diagnosis Date'], format='%d-%m-%Y')
```

```python
# Changing columns name to useful names
rename_col= {'OCD Diagnosis Date':'Diagnosis Date','Duration of Symptoms (months)':'Symptoms Duration(months)','Family History of OCD':'Family History',
             'Y-BOCS Score (Obsessions)':'Obsessions Score','Y-BOCS Score (Compulsions)':'Compulsions Score'}

ocd.rename(columns=rename_col,inplace=True)
```

```python
# Rechecking the dataset

ocd.head(10)
```

```python
# Rechecking info after cleaning
ocd.info()
```

# 3. Exploratory Data Analysis

### 3.1 Visualizing Demographic Data

```python
# Age distribution of patients

plt.figure(figsize=(15,5))
sns.histplot(ocd['Age'],bins=40,kde=True,color='teal')
plt.title('Age Distribution of Patients')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

* From the above graph it can be derived that young population around the age of 35 and below are diagnosed with OCD. It shows young ones are having mental disorders compared to people above the age of 35.

```python
# Gender distribution

gender_dist=ocd['Gender'].value_counts()
gender_dist
```

```python
# Plotting the Gender distribution

plt.figure(figsize=(8,5))
sns.barplot(gender_dist,color='teal')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

* There are almost equal number of male and female patients of OCD.

```python
# Ethnicity distribution of patients

ethnic=ocd['Ethnicity'].value_counts()
ethnic
```

```python
# Plotting the Ethnicity distribution of patients

plt.figure(figsize=(15,5))
sns.countplot(ocd['Ethnicity'],color='teal')
plt.title('Ethnicity Distribution')
plt.xlabel('Count')
plt.ylabel('Ethnicity')
plt.show()
```

* Hispanic,Asian and Caucasian have most patients of OCD than African.

### 3.2 Clinical Data Analysis

```python
# Analysing minimum and maximum duration of symptoms

print('Minimum duration of symptoms patients had : ',ocd['Symptoms Duration(months)'].min(),'months')
print('')
print('Maximum duration of symptoms patients had : ',ocd['Symptoms Duration(months)'].max(),'months')
```

* From the above analysis it is found that patients have OCD symptoms from 6 months. And some patients have symptoms which lasts for 240 months. This is huge time for symptoms.

```python
# Distribution of symptom duration

plt.figure(figsize=(15,5))
sns.histplot(ocd['Symptoms Duration(months)'],bins=20,kde=True,color='teal')
plt.title('Distribution of Symptom Duration')
plt.xlabel('Duration(in months)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Boxplot of Y-BOCS Scores by Gender

plt.figure(figsize=(10,8))

plt.subplot(1,2,1)
sns.boxplot(x='Gender',y='Obsessions Score',data=ocd,color='teal')
plt.title(' Obsession Scores by Gender')


plt.subplot(1,2,2)
sns.boxplot(x='Gender',y='Compulsions Score',data=ocd,color='teal')
plt.title('Compulsions Scores by Gender')
plt.show()
```

* The severity of instrusive thoughts that is Obsessions Score is between 10 and 32 for female and between 10 and 28 for male.

* The severity of repetitive behaviors that is Compulsions Score is between 09 and 28 both for female and  male.

```python
# Relationship between Obsession and Compulsion Y-BOCS Scores

plt.figure(figsize=(8,8))
sns.scatterplot(x='Obsessions Score',y='Compulsions Score',hue='Gender',data=ocd)
plt.title('Relationship between Obsession and Compulsion Y-BOCS Scores')
plt.xlabel('Obsessions Score')
plt.ylabel('Compulsions Score')
plt.show()
```

* All the patients have almost equal Y-BOCS Score.

```python
# Correlation matrix among age,Duration of Symptoms,Y-BOCS score

corr_matrix = ocd[['Age','Symptoms Duration(months)','Obsessions Score','Compulsions Score']].corr()
corr_matrix
```

```python
# Plotting the co-relation in heatmap

plt.figure(figsize=(8,5))
sns.heatmap(corr_matrix,annot=True,fmt=".2f",cmap='RdBu')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Adding a column after Y-BOCS Score (Compulsions)

col_index = ocd.columns.get_loc('Compulsions Score')
ocd.insert(col_index+1,'Total Y-BOCS Score',ocd['Obsessions Score']+ocd['Compulsions Score'])
```

```python
# Plotting total YBOCS score
sns.boxplot(x='Total Y-BOCS Score',data=ocd,color='teal')
plt.show()
sns.histplot(x='Total Y-BOCS Score',data=ocd,color='teal',kde=True)
plt.show()
```

* From the above graph it is evident that Total Y-BOCS score which should not be exceeding 40,here it is exceeding 40.
* It shows there is some wrong inputs in Obsessions Score column and Compulsions Score column as each column score must not exceed 20.
